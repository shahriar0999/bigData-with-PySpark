```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType

schema = StructType([
    StructField("id", StringType(), True),
    StructField("customer_name", StringType(), True),
    StructField("my_product", StringType(), True),
    StructField("quantity", IntegerType(), True),
    StructField("price", DoubleType(), True),
    StructField("remarks", StringType(), True)
])

df = spark.read\
    .schema(schema)\
    .csv("/Volumes/readcsv/mylocal_csv_file/local_to_cloud/dataDIR/sales_no_header.csv")

display(df)
```

```python
# alternative code
schema = StructType([
    StructField("id", StringType(), True),
    StructField("customer_name", StringType(), True),
    StructField("my_product", StringType(), True),
    StructField("quantity", IntegerType(), True),
    StructField("price", DoubleType(), True),
    StructField("remarks", StringType(), True)
])

df3 = spark.read.csv("/Volumes/readcsv/mylocal_csv_file/local_to_cloud/dataDIR/sales_no_header.csv", schema=schema)

display(df3)
```
